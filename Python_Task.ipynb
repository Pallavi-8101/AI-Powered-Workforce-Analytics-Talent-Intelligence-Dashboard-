### Smart File Organizer Project

This project demonstrates how to organize files into categorized folders based on their extensions using Python's `os` and `shutil` modules.

```python
import os
import shutil

# Define the base directory for the project
base_dir = '/content/TestFiles'

# 1. Create the main project folder if it doesn't exist
if os.path.exists(base_dir):
    shutil.rmtree(base_dir) # Remove existing folder to ensure a clean run
os.makedirs(base_dir, exist_ok=True)
print(f"Created main project folder: {base_dir}")
```

```python
# 2. Create sample files inside the main folder
sample_files = {
    'photo.jpg': b'JPEG image data',
    'document.pdf': b'%PDF-1.4 sample document',
    'song.mp3': b'ID3 audio sample',
    'video.mp4': b'MPEG-4 video',
    'archive.zip': b'PK\x03\x04 zip archive',
    'notes.txt': 'This is a text note.'.encode('utf-8'),
    'presentation.pptx': b'PowerPoint presentation',
    'spreadsheet.xlsx': b'Excel spreadsheet',
    'another_photo.jpg': b'Another JPEG image data',
    'report.pdf': b'%PDF-1.4 another document',
    'README': 'This is a README file without an extension.'.encode('utf-8')
}

print("\nCreating sample files...")
for filename, content in sample_files.items():
    file_path = os.path.join(base_dir, filename)
    with open(file_path, 'wb') as f:
        f.write(content)
    print(f"  Created: {filename}")
```

```python
# 3. Define category folders and their corresponding extensions
category_map = {
    'Images': ['.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff'],
    'Videos': ['.mp4', '.mkv', '.avi', '.mov', '.wmv', '.flv'],
    'Documents': ['.pdf', '.doc', '.docx', '.ppt', '.pptx', '.xls', '.xlsx', '.txt'],
    'Audio': ['.mp3', '.wav', '.aac', '.flac', '.ogg'],
    'Archives': ['.zip', '.rar', '.7z', '.tar', '.gz'],
    'Others': [] # Files with unrecognized extensions will go here
}

# Create category folders
for category_name in category_map.keys():
    category_path = os.path.join(base_dir, category_name)
    os.makedirs(category_path, exist_ok=True)
    print(f"  Created category folder: {category_name}")
```

```python
# 4. Organize files
print("\nOrganizing files...")

# Initialize a dictionary to store the count of files moved to each category
files_moved_summary = {category: 0 for category in category_map.keys()}
files_moved_summary['Uncategorized'] = 0

for item in os.listdir(base_dir):
    item_path = os.path.join(base_dir, item)

    # Skip directories (category folders) during organization
    if os.path.isdir(item_path):
        if item not in category_map.keys(): # Only print for non-category folders
            print(f"  Skipping directory: {item}")
        continue

    # Get file extension
    file_name, file_extension = os.path.splitext(item)
    file_extension = file_extension.lower()

    destination_folder = None
    for category, extensions in category_map.items():
        if file_extension in extensions:
            destination_folder = category
            break

    # If no specific category is found, move to 'Others'
    if destination_folder is None:
        destination_folder = 'Others'

    destination_path = os.path.join(base_dir, destination_folder, item)

    # Handle duplicate file names safely
    original_destination_path = destination_path
    counter = 1
    while os.path.exists(destination_path):
        name, ext = os.path.splitext(item)
        destination_path = os.path.join(
            base_dir, destination_folder, f"{name}({counter}){ext}"
        )
        counter += 1

    shutil.move(item_path, destination_path)
    print(f"  Moved '{item}' to '{destination_folder}' as '{os.path.basename(destination_path)}'")
    files_moved_summary[destination_folder] += 1
```

```python
# 5. Print a summary of files moved
print("\n--- Organization Summary ---")
for category, count in files_moved_summary.items():
    print(f"  {category}: {count} file(s)")
```

```python
# 6. Display the final folder structure
print("\n--- Final Folder Structure ---")
def print_folder_structure(startpath):
    for root, dirs, files in os.walk(startpath):
        level = root.replace(startpath, '').count(os.sep)
        indent = ' ' * 4 * (level)
        print(f'{indent}{os.path.basename(root)}/')
        subindent = ' ' * 4 * (level + 1)
        for f in files:
            print(f'{subindent}{f}')

print_folder_structure(base_dir)
```

```python
# 7. Print completion message
print("\nProject completed successfully!")
print("This project demonstrates Python file handling and automation skills.")
```
